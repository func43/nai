```python
from itertools import permutations

letters = "CROSADNGE"

for perm in permutations("0123456789", len(letters)):
    mapping = dict(zip(letters, perm))

    if mapping['C'] == '0' or mapping['R'] == '0' or mapping['D'] == '0':
        continue

    CROSS  = int(mapping['C'] + mapping['R'] + mapping['O'] + mapping['S'] + mapping['S'])
    ROADS  = int(mapping['R'] + mapping['O'] + mapping['A'] + mapping['D'] + mapping['S'])
    DANGER = int(mapping['D'] + mapping['A'] + mapping['N'] + mapping['G'] + mapping['E'] + mapping['R'])

    if CROSS + ROADS == DANGER:
        print("Solution:")
        print("CROSS =", CROSS)
        print("ROADS =", ROADS)
        print("DANGER =", DANGER)
        print("Mapping =", mapping)
        break
```
